```python
def save_atlas(df,atlas_name):
    all_identifications = []
    for x in df.index:
        myID = metob.CompoundIdentification()
        compound_label = df['compound_identification name'][x] #if no name, then use label as descriptor
        if (df.inchi_key[x] != '') & (type(df.inchi_key[x]) != float): # this logic is where an identified metabolite has been specified
            c = metob.retrieve('Compounds',neutralized_inchi_key=df.inchi_key[x],username = '*') #currently, all copies of the molecule are returned.  The 0 is the most recent one. 
            if c:
                c = c[0]
        else:
            c = 'use_label'
 
        if c:
            if c != 'use_label':
                myID.compound = [c]
            myID.name = compound_label

            mzRef = metob.MzReference()
            mzRef.mz = df['mz_ref mz'][x]
            mzRef.mz_tolerance = df['mz_ref mz_tolerance'][x]
            mzRef.mz_tolerance_units = 'ppm'
            mzRef.detected_polarity = df['mz_ref detected_polarity'][x]
            myID.mz_references = [mzRef]

            rtRef = metob.RtReference()
            rtRef.rt_units = 'min'
            rtRef.rt_min = df['rt_ref rt_min'][x]
            rtRef.rt_max = df['rt_ref rt_max'][x]
            rtRef.rt_peak = df['rt_ref rt_peak'][x]
            myID.rt_references = [rtRef]
                    
            all_identifications.append(myID)

    myAtlas = metob.Atlas()
    myAtlas.name = atlas_name
    myAtlas.compound_identifications = all_identifications
    metob.store(myAtlas)
    return myAtlas
```

```python


unique_atlases = df['atlas name'].unique()
for u_atlas in unique_atlases:
    last_timestamp = sorted(df[(df['atlas name'] == u_atlas)]['compound_identification last_modified'].unique())[-1]
    atlas_to_save = df[(df['atlas name'] == u_atlas) & (df['compound_identification last_modified'] == last_timestamp)]
    for polarity in atlas_to_save['mz_ref detected_polarity'].unique():
        atlas = save_atlas(atlas_to_save[atlas_to_save['mz_ref detected_polarity']==polarity],u_atlas)

    
```
